```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.groupby('Species').describe()
```

```python
print(df.columns)
```

```python
setosa = df[df['Species']=='Iris-setosa']['SepalLengthCm'].tolist()
versicolor = df[df['Species']=='Iris-versicolor']['SepalWidthCm'].tolist()
virginica = df[df['Species']=='Iris-virginica']['PetalLengthCm'].tolist()

print(setosa,versicolor,virginica)
```

```python
df.groupby('Species').agg({

    'SepalLengthCm':['mean','min','max'],
    'SepalWidthCm':['mean','min'],
    'PetalLengthCm':['mean','min',lambda x : x.mode()[0]]
    
})
```
