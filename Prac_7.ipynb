```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
text = 'Data analytics is the process of extracting usefull information from the row data'
```

```python
tokens = word_tokenize(text)
```

```python
print(tokens)
```

```python
tags = pos_tag(tokens)
```

```python
print(tags)
```

```python
stop_words = set(stopwords.words('english'))
filter_tokens = [x for x in tokens if x.lower() not in stop_words]
print(filter_tokens)
```

```python
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filter_tokens]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmitized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]
print(lemmitized_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "Text analytics is useful",
    "Text data analysis is important",
    "Analytics extracts information from text"
]
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
feature_names = vectorizer.get_feature_names_out()
```

```python
tfidf_array = tfidf_matrix.toarray()
```

```python
print("Feature Names:", feature_names)
print("TF-IDF Matrix:\n", tfidf_array)
```
